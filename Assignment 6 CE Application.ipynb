# Explicit method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.25
k = 0.025
x = np.arange(0, 1+h, h)
t = np.arange(0, 0.1+k, k)
boundaryConditions =[0, 0]
initialConditions = np.sin(np.pi*x)

n = len(x)
m = len(t)
T = np.zeros((n,m))
T[0,:] = boundaryConditions[0]
T[-1, :] = boundaryConditions[1]
T[:, 0] = initialConditions

factor = k/h**2
for j in range(1, m):
    for i in range(1, n-1):
        T[i,j] = factor*T[i-1, j-1] + (1-2*factor)*T[i, j-1] + factor*T[i+1, j-1]  #Expression used
T.round(3)



```

```python
for j in range(m):
    plt.plot(x, T[:,j])
plt.legend(t)
plt.xlabel('distance [m]')
plt.ylabel('Temperature [degree C]')
```
