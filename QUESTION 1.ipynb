```python
#Import necessary library
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

y0 = 1 #Initial condition

def dydt(y,t):
    dydt = -2*y*t    #Differential Equation
    
    return dydt
#Interval
start = 0
end = 2

t = np.linspace(start, end, end)

yinitial = np.linspace(y0,y0,end)

y = sc.integrate.odeint(dydt, y0, t)  #Integration of the differential equation

#Plotting the results
plt.figure()
plt.plot(t, y)
plt.xlabel('y(t)')
plt.ylabel('t')
plt.show()
```
